Write a Python program to simulate the calculation of π using Monte Carlo method. Your program draws the circle and the square, and plot the random points in the graph as shown in Figure 1. You need simulate diﬀerent numbers of points (at least include 10, 500, 1,000, 5,000, 10,000, 50,000, 80,000, 100,000, 200,000) and calculate the value π for each diﬀerent number of points. Then draw a line chart to show the change of the accuracy of the value of π calculated when the number of points is increased. Submit Python code (2 points, no point will be given for a program that doesn’t correctly implement the simulation algorithm), and simulation results including the graph plots (1 point), π values (1 point), and line chart (1 point).


```
import matplotlib as plt
import numpy as np

def generate_uniform_points(size):
  return np.random.uniform(0,1,size)
  pass

def generate_randomList(*args):
  return (args[0]-0.5)**2+(args[1]-0.5)**2<(1/4)
  pass

def visualize_plot(*args):
  plt.pyplot.show(block = False)
  actual_plot = plt.pyplot.figure(figsize=(5,5)).add_subplot(1,1,1)
  rectangle = plt.pyplot.Rectangle((0,0),1,1,edgecolor='r',fill=False)
  circle = plt.pyplot.Circle((0.5,0.5),radius=0.5,edgecolor='b',fill=False)
  actual_plot.add_patch(rectangle)
  actual_plot.add_patch(circle)
  plt.pyplot.scatter(args[0],args[1],s=1,color='r')
  plt.pyplot.scatter(args[0],args[1],s=1,color='b')
  pass

def generate_List(size):
  x,y = generate_uniform_points(size),generate_uniform_points(size)
  result = generate_randomList(x,y)
  x_plot,y_plot = x[result],y[result]
  visualize_plot(x_plot,y_plot)
  return result
  pass

def computeTheValueOfPi(plotters,size):
  pi_val = (list(plotters).count(True)/size)*4
  print(f"Value of pi is : {pi_val}\n")
  pass

def executor():
  inputCount = [10,500,1000,5000,10000,50000,80000,100000,200000]
  for i in inputCount:
    arrayOfPoints = generate_List(i)
    computeTheValueOfPi(arrayOfPoints,i)
  pass

executor()
```
